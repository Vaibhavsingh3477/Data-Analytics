### Lambda Functions

```python
addition1 = lambda x,y,z : x+y+z
addition2 = lambda x,y : x+y

print(addition1(1,2,3))
print(addition2(1,2))
```

##### Without Lambda

```python
numbers = [1,2,3,4,5]

def square(x):
    return x**2

print(square(2));
print(square(3));
```

#### With Lambda ~ Map

```python
numbers = [1,2,3,4,5]
squared = list(map(lambda x: x**2, numbers))

print(squared)
```
