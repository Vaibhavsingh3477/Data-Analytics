## Map Function 

```python
def square(x):
    return x * x

print(square(2))
print(square(3))
```

```python
numbers=[1,2,3,4,5]
map(square, numbers)
list(map(square, numbers))
# Output: [1, 4, 9, 16, 25] 
```

#### Lambda function with map   

```python
numbers = [1, 2, 3, 4, 5]
list(map(lambda x: x * x, numbers))
# Output: [1, 4, 9, 16, 25]
```

```python
number1 = [1, 2, 3]
number2 = [4, 5, 6]

added_numbers=list(map(lambda x,y:x+y,number1,number2))
print(added_numbers)
# Output: [5, 7, 9]
```

```python
## map() to convert a list of strings to integers
# Use map to convert strings to integers

string_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, string_numbers))    

print(int_numbers)
```

```python
words=['apple', 'banana', 'cherry']
upper_words=list(map(str.upper,words))
print(upper_words)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35}
]

list(map(get_name, people))
```

## Conclusion: Map Function

The `map()` function is a powerful built-in Python function that applies a specified function to every item in an iterable (like a list) and returns a map object (iterator). 

### Key Takeaways:
- **Syntax**: `map(function, iterable)`
- Returns an iterator, so use `list()` to convert to a list
- Works with built-in functions like `int`, `str.upper`, etc.
- Can use lambda functions for inline function definitions
- Can handle multiple iterables (they must be the same length)

### Use Cases:
1. Transforming data (converting types, applying transformations)
2. Applying mathematical operations to lists
3. Converting collections of strings to other types
4. Applying custom functions to elements

The map function provides a clean, functional approach to data transformation without explicit loops.
